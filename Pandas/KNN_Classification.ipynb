```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
soc=pd.read_csv("../Datasets/Social_Network_Ads.csv")
soc
```

```python
X=soc.iloc[:,[2,3]].values
X
```

```python
y=soc.iloc[:,4].values
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler 
sc= StandardScaler()
X_train = sc.fit_transform(X_train)
X_test=sc.transform(X_test)
# To make it in similiar Ranges we have to scale our Inputs 
```

```python
X_train
```

```python
y_train
```

```python
len(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
cf=KNeighborsClassifier(n_neighbors=5)
# to check upto 5  neighbors:
cf.fit(X_train,y_train)
pred=cf.predict(X_test)
pred
```

```python
y_test
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test,pred))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,pred)
cm
```

```python
# Checking Error at different values of K 
```

```python
error1=[]
error2=[]
for k in range(1,10):
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train)
    y_pred1=knn.predict(X_train)
    error1.append(np.mean(y_train!=y_pred1))
    y_pred2=knn.predict(X_test)
    error2.append(np.mean(y_test!=y_pred2))
```

```python
plt.plot(range(1,10),error1,label='train')
plt.plot(range(1,10),error2,label='test')
plt.xlabel('K-value')
plt.ylabel('Error')
plt.legend()
# to show the Specification and pointers in the Graph:
plt.show()
```
