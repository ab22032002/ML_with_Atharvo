```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris=datasets.load_iris()
iris
```

```python
iris_data = pd.DataFrame(iris.data)
iris_data.columns=iris.feature_names
iris_data['Flower_Type']= iris.target
iris_data.head()
```

```python
iris_X=iris_data.iloc[:,[0,1,2,3]].values
iris_y=iris_data.iloc[:,4].values
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
plt.scatter(iris_X[iris_y==0,0],iris_X[iris_y==0,1],s=100,c='blue',label='Type1')
# sepal length(0) and sepal width(1) the plotting is done
plt.scatter(iris_X[iris_y==1,0],iris_X[iris_y==1,1],s=100,c='yellow',label='Type2')
plt.scatter(iris_X[iris_y==2,0],iris_X[iris_y==2,1],s=100,c='green',label='Type3')
plt.legend()
plt.show()
```

```python
import scipy.cluster.hierarchy as hc
plt.figure(figsize=(20,7))
plt.title('DendoGrams:')
hc.dendrogram(hc.linkage(iris_X,method='centroid'))
plt.title("Dendrograms")
plt.xlabel("SAMple  index")
plt.ylabel('generally:')
plt.show()
```

```python
import scipy.cluster.hierarchy as hc
plt.figure(figsize=(20,7))
plt.title('DendoGrams:')
hc.dendrogram(hc.linkage(iris_X,method='complete'))
plt.title("Dendrograms")
plt.xlabel("SAMple  index")
plt.ylabel('generally:')
plt.show()
```

```python
# method_ ward so the variance b/w the datapoints is minimized
import scipy.cluster.hierarchy as hc
plt.figure(figsize=(20,7))
plt.title('DendoGrams:')
hc.dendrogram(hc.linkage(iris_X,method='ward'))
plt.title("Dendrograms")
plt.xlabel("SAMple  index")
plt.ylabel('generally:')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster= AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
cluster.fit(iris_X)
labels=cluster.labels_
labels
```

```python
plt.figure(figsize=(10,7))
plt.scatter(iris_X[labels==0,0],iris_X[labels==0,1],s=100,c='blue',label='Type1')
plt.scatter(iris_X[labels==1,0],iris_X[labels==1,1],s=100,c='yellow',label='Type2')
plt.scatter(iris_X[labels==2,0],iris_X[labels==2,1],s=100,c='green',label='Type3')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.scatter(iris_X[labels==0,2],iris_X[labels==0,3],s=100,c='blue',label='Type1')
plt.scatter(iris_X[labels==1,2],iris_X[labels==1,3],s=100,c='yellow',label='Type2')
plt.scatter(iris_X[labels==2,2],iris_X[labels==2,3],s=100,c='green',label='Type3')
plt.legend()
plt.show()
# petal length and petal width:
```
