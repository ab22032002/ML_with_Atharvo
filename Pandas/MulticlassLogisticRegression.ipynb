```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
iris
```

```python
# Iris is a kind of flower
# we have to predict the species of the flower by analysing its petals and shape size etc
```

```python
iris.target
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
iris_data=pd.DataFrame(iris.data,columns=iris.feature_names)
iris_data
```

```python
iris_data['species']=iris.target
iris_data
```

```python
iris_data['species_name']=iris_data.species.apply(lambda x:iris.target_names[x])
iris_data
```

```python
y=iris_data['species']
y
```

```python
X=iris_data.drop(['species','species_name'],axis=1)
X
```

```python
sns.pairplot(iris_data)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=19)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
model.score(X_test,y_test)
# Score closer to 1 thats the accuracy of the model
```

```python
model.predict(iris.data[0:65])
```

```python
iris.target[0:65]
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
sns.heatmap(cm,annot=True)
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

# Digits DataSet

```python
from sklearn.datasets import load_digits
digi=load_digits()
digi
```

```python
plt.gray()
```

```python
for i in range(5):
    plt.matshow(digi.images[i])
```

```python
digidata=pd.DataFrame(digi.data,columns=digi.feature_names)
digidata
```

```python
digidata['out']=digi.target
digidata['out']
```

```python
X=digidata.drop('out',axis='columns')
X
```

```python
y=digidata.out
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression
dmodel=LogisticRegression()
dmodel.fit(X_train,y_train)
```

```python
y_pred=dmodel.predict(X_test)
```

```python
dmodel.score(X_test,y_test)
```

```python
dmodel.predict(digi.data[0:5])
```

```python
digi.target[0:5]
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
sns.heatmap(cm,annot=True)
```
