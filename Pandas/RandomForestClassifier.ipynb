```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
social=pd.read_csv("../Datasets/Social_Network_Ads.csv")
social
```

# Spliting Output Data And input Data

```python
X=social.iloc[:,[2,3]].values
X
```

```python
y=social.iloc[:,4].values
```

```python
y
```

```python
# Generating Testing and Training Data
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler 
sc= StandardScaler()
X_train = sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_test
```

```python
from sklearn.ensemble import RandomForestClassifier
cfr=RandomForestClassifier(n_estimators=5,random_state=3)
cfr.fit(X_train,y_train)
pred=cfr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,pred)
cm
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test,pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred))
```
