```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
mc=pd.read_csv("../Datasets/Mall_Customers.csv")
mc
```

```python
X=mc.iloc[:,[3,4]].values
X
```

```python
from sklearn.cluster import KMeans
fig=plt.figure(figsize=(9,5))
wcss=[]
for i in range(1,11):
    cf=KMeans(n_clusters=i,init='k-means++')
    cf.fit(X)
    wcss.append(cf.inertia_)

plt.plot(range(1,11),wcss,marker='.')
plt.title('Elbow Method')
plt.xlabel('no of Clusters:')
plt.ylabel('wcss')
plt.show()
```

```python
# this method gives value of k equals to 5 s o n_clusters =5
```

```python
cf=KMeans(n_clusters=5,init='k-means++')
y_kmeans=cf.fit_predict(X)
y_kmeans
```

```python
fig=plt.figure(figsize=(10,5))
plt.scatter(X[y_kmeans==0,0],X[y_kmeans==0,1],color='red',marker='.',label='Cluster1')
plt.scatter(X[y_kmeans==1,0],X[y_kmeans==1,1],color='yellow',marker='.',label='Cluster2')
plt.scatter(X[y_kmeans==2,0],X[y_kmeans==2,1],color='black',marker='.',label='Cluster3')
plt.scatter(X[y_kmeans==3,0],X[y_kmeans==3,1],color='blue',marker='.',label='Cluster4')
plt.scatter(X[y_kmeans==4,0],X[y_kmeans==4,1],color='green',marker='.',label='Cluster5')
plt.scatter(cf.cluster_centers_[:,0],cf.cluster_centers_[:,1],marker='x',c='purple',label='centroid')
plt.title('Clustering Using Kmeans')
plt.xlabel('x__')
plt.ylabel('y__')
plt.show()
```
