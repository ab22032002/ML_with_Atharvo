```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
# Considering Statestical Columns Only
# Segragating the input and output into 2 variables
```

```python
X=social.iloc[:,[2,3]].values
y=social.iloc[:,4].values
X
```

```python
y
```

```python
sns.heatmap(social.corr(),annot=True)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=2)
X_train
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train
# Applying while the datasets have huge differences in range of values
```

```python
X_test
```

```python
from sklearn.tree import DecisionTreeClassifier
cf=DecisionTreeClassifier(random_state=6)
cf.fit(X_train,y_train)
```

```python
y_pred=cf.predict(X_test)
y_pred
```

```python
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
print('Accuracy of the model is :{}'.format(accuracy_score(y_test,y_pred)*100))
```

```python
accuracy=[]
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

for i in range(1,10):
    model = DecisionTreeClassifier(max_depth = i,random_state=0)
    model.fit(X_train,y_train)
    pred=model.predict(X_test)
    score=accuracy_score(y_test,pred)
    accuracy.append(score)
    
    
plt.figure(figsize=(12,6))
plt.plot(range(1,10),accuracy,color='red',linestyle='dashed',marker='o',markerfacecolor='blue',markersize=10)
plt.title('Finding Best Max_depth')
plt.xlabel('pred')
plt.ylabel('score')
```

```python
from sklearn.tree import DecisionTreeClassifier
cff=DecisionTreeClassifier(max_depth=2,random_state=0)
cff.fit(X_train,y_train)
pred=cff.predict(X_test)
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test,pred))
```
